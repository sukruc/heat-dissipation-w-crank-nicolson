## Visualizing Heat Dissipation w/OLS

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import sys

from heatutils import initialize, calculate, propagate, heatmap
```

```python
metadata = pickle.load(open('calculated_data/test.pk','rb'))
```

```python
X = metadata['X']
Y = metadata['Y']
dx = metadata['dx']
dt = metadata['dt']
T = metadata['data']
```

```python
cm = plt.cm.get_cmap('jet')
fig, ax = plt.subplots(2,2, figsize=(12,8), sharex=True, sharey=True)
heatmap(X, Y, T, dx, dt, 300, ax[0,0])
heatmap(X, Y, T, dx, dt, 900, ax[0,1])
heatmap(X, Y, T, dx, dt, 1800, ax[1,0])
heatmap(X, Y, T, dx, dt, 3600, ax[1,1])
```
