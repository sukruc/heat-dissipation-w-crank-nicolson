## Solution of Crank-Nicolson Finite Difference Scheme with CG method

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import sys

from heatutils import initialize, calculate, propagate, heatmap
```

```python
metadata = pickle.load(open('calculated_data/test.pk','rb'))
```

```python
X = metadata['X']
Y = metadata['Y']
dx = metadata['dx']
dt = metadata['dt']
T = metadata['data']
```

```python
cm = plt.cm.get_cmap('jet')
fig, ax = plt.subplots(2,2, figsize=(12,8), sharex=True, sharey=True)
time_viz = iter([300, 900, 1800, 3600])
for i in range(2):
    for j in range(2):
        heatmap(X, Y, T, dx, dt, next(time_viz), ax[i,j])
```
